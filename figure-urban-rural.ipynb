```python
thresholds = np.repeat(np.arange(9),5)
metrics = np.tile(np.array([
    'balanced accuracy',
    'F1',
    'precision',
    '1-recall',
    '0-recall'
]), 9)
classes = np.array(['unsettled', 'rural', 'urban'])

index = pd.MultiIndex.from_arrays([thresholds, metrics], names=('threshold', 'metric'))

df = pd.DataFrame(index=index, data=data.reshape(-1,3), columns=classes)
df
```

```python
x = np.arange(3)
width = 0.2

fig, ax = plt.subplots(figsize=(7.4, 5.2))
rects1 = ax.bar(x - 8/7*width, df.loc[0, 'balanced accuracy'].to_numpy().ravel(), width, label='threshold = 0', edgecolor='black')
rects2 = ax.bar(x, df.loc[3, 'balanced accuracy'].to_numpy().ravel(), width, label='threshold = 3', edgecolor='black')
rects3 = ax.bar(x + 8/7*width, df.loc[6, 'balanced accuracy'].to_numpy().ravel(), width, label='threshold = 6', edgecolor='black')

ax.set_ylabel('balanced accuracy')
ax.set_title('Balanced accuracy across settlement classes and thresholds')
ax.set_xticks(x)
ax.set_xticklabels(['unsettled', 'rural', 'urban'])
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(np.around(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```

```python
# https://riptutorial.com/matplotlib/example/10473/single-legend-shared-across-multiple-subplots

fig, axes = plt.subplots(2, 2, sharex=True, figsize=(10,4))

for ax, metric in zip(axes.ravel(), np.unique(metrics)):
    l1 = ax.plot(df.xs(metric, level='metric')['rural'], marker='+', label='rural')
    l2 = ax.plot(df.xs(metric, level='metric')['urban'], marker='o', label='urban')
    ax.set_ylabel(metric)
    ax.set_xticks(range(9))
    ax.set_xticklabels(range(1,10))

axes[1][0].set_xlabel('threshold')
axes[1][1].set_xlabel('threshold')
fig.legend([l1,l2],
           labels=['rural','urban'],
           loc='center right',
           borderaxespad=0.1,
           title='')
#fig.tight_layout()
#fig.suptitle('Balanced accuracy across settlement classes and thresholds')
plt.subplots_adjust(right=0.90)
plt.show()
```
